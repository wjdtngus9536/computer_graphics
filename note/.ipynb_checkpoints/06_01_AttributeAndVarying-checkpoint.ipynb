연립 1차 방정식
중간에 있는 스칼라 값들을 어떤 직사각형 형태로 숫자들을 모아 놓은 것을 행렬이라고 얘기 함.
연립 방정식을 길게 쓰는 것 보다 행렬로 표현 하는 것이 더 간단하기 때문에 행렬을 쓰는 경우가 많다.

행렬을 실제로 의미하는 연립 방정식으로 복원하려면 곱할 필요가 있다.


row, column, vector
- m x n matrix의 m row, n column
- 행-벡터(row vector) = 1 x n 행렬
- 열-벡터 = m x 1 행렬


vector-vector 곱셈
a = (ax, ay, az)
b = (bx, by, bz)

벡터를 곱하는 방법 2가지
1. dot product(스칼라곱, 내적) : a * b = axbx + ayby + azbz
2. cross product(벡터곱, 외적) :
    - 일종의 행렬식(determinant, 결정요인) 계산 : 단, i,j,k 는 basis vector 
    
    외적은 벡터와 벡터를 곱해서 새로운 벡터가 나오는 과정
    강의를 진행하며 더 자세히 알아보자
    
행렬과 벡터의 관계
- 행렬은 여러 개의 row-vector로 해석 가능
- 행렬에 벡터를 곱해서 벡터를 구하는 새로운 연산

행렬과 행렬의 곱셈



3D, 4D 행렬식
행렬과 벡터의 곱연산을 하는 이유 : **좌표를 변환**시키는 게 가능하기 때문



## 선형 보간 linear Interpolation

수치 예측법
- 자료에 기반한 예측법
    - interpolation 보간법, 내삽법
        - 주어진 자료의 내부 구간을 예측 (정확도 높음)
    - extrapolation 보외법, 외삽법
        - 주어진 자료의 외부 구간을 예측 (정확도가 많이 떨어짐)
- 예측에 사용하는 함수
    - linear interpolation : 1차 함수 (직선) 사용
    - cubic interpolation : 3차 함수 (곡선) 사용
    

선형 보간
