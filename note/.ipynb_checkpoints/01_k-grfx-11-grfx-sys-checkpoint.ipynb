## 2. 래스터 시스템 Rester System

### 그래픽스 시스템
 - 이미지는 `프레임버퍼(화면에 나오는 부분을 그대로 저장하는 것들)`에 저장 
 
### 래스터 시스템(이미지 저장 방법)
 - 이미지 -> (색깔 있는) 사각형의 2차원 배열 = 2D array of colored rectangles
 - 픽셀 pixel = picture element = a single colored rectangle in the image

### 래스터 시스템 용어들
 - 픽셀 pixel = picture element
 - 래스터 raster = `픽셀의 2차원 배열` a 2D array of pixels
 - 프레임 버퍼 framebuffer(래스터 방식으로 저장된 것 중에 가장 중요한 것)  = 픽셀을 저장한 `메모리 영역`
     - 우리가 쓰는 그래픽 카드에는 반드시 프레임버퍼가 들어가 있음.
     
### 프레임버퍼의 기능
 - 프레임버퍼 -> 화면으로 **1:1 대응 관계**
 - 우리는 프레임 버퍼의 2차원 배열에 있는 내용을 바꿀 것.
 - 이게 하드웨어 적으로 구현됨

## 합성 카메라 모델(Synthetic Camera Model)

바늘구멍 카메라의 수학적 해석
- 가장 간단한 카메라 모델
- 인간 시각 시스템의 대안
- 비례식 해석
원점에서 물체까지 거리 : 원점에서 필름까지 거리 = 원점에서 물체까지 x좌표 : 원점에서 필름까지 x좌표

컴퓨터 그래픽스 카메라 모델
- 바늘구멍 카메라 모델의 변형

    - 윈도우 이동 가능 : 윈도우 clipping

## 그래픽스 아키텍처

1. 모델링 - 렌더링 패러다임
    - 가장 간단한 그래픽스 아키텍처, 가장 오래된 역사
    - 서로 독립적인 프로그램, 파일로 자료 교환
    - 모델링 프로그램 / 렌더링 프로그램이 서로 독립적
    - e.g., 적용 예 : Pixar Renderman Interface (.rib)
    
2. CPU 단독 처리 모델
    - 문제점 : CPU가 해야할 계산량이 너무 많다.
    
3. DPU(Display Processing Unit) 모델
    - 그래픽스 처리만 전담하는, 특별한 목적의 프로세퍼
    - 파이프라인 모델로 가는 중간 단계
    
4. 파이프라인 모델
    - 전용 그래픽 카드 방식 모델
        - 그래픽스 처리 과정을 단계별로 `VLSI칩`으로 처리
        - 실리콘 그래픽스가 도입
            0. 호스트 컴퓨터
            1. 각 정점에 대해 여러 변환을 수행
            2. primitive 조립
            3. rasterizer
            4. fragment processing
            5. 모니터
            
            
5. 프로그래머블 파이프라인 모델  
고정된 VLSI칩 대신, 프로그래밍 가능한 processor를 사용
Shader 는 vertex 정보들을 화면에 보여질수 있도록 변환하는 작업을 하는 프로그램(명령어 집합)  
스크린 위에 낱개의 픽셀마다 실행되며 이 실행들이 한번에 일어납니다. 이러한 작업은 GPU 에서 일어납니다.


- vertex shader : 프로그래밍 가능한 정점 처리기
- fragment(파편) shader : Fragment Shader 는 각각의 픽셀들이 어떤 색상으로 그려질지를 정의


6. GPU(Graphic Orocessing Unit) : 프로그래머블 파이프라인에 사용되는 그래픽스 전용 프로세서
    - 빠른처리가 필요 >> floating point 연산, 4x4 행렬 연산에 최적화

- GPGPU(General Purpose - GPU)
    - GPU를 일반적 목적으로 사용
    - 즉, GPU를 CPU대신 계산에 사용
    - 대규모 병렬 컴퓨팅 MPC(massively parallel computing)의 시작
