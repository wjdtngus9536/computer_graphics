## OpenGL 라이브러리
c/cpp 컴파일 과정
- 컴파일 : 소스코드 >> 목적파일
- 링크 : 목적파일 + (정적) 라이브러리 파일 >> 실행 파일
- 실행 : 실행 파일 + DLL(Dynamic-link library) 파일

<br>

c/cpp 라이브러리
- 목적 : 자주 쓰이는 기능을 잘 구현해서 재사용 하기 위해
    - 구현 : 시스템 프로그래머
    - 사용 : 응용 프로그래머
    
- API : 응용 프로그램 인터페이스
    - 인터페이스 규약, 어떻게하면 저 함수가 불리는지를 알고 있음
    
- 라이브러리 : 구현된 함수들을 미리 컴파일 해서 모아둔 파일
    - API의 실제 구현, 경우에 따라 1개의 API를 여러 개의 라이브러리로 구현
    
- 헤더 파일 : (컴파일러를 위한) 라이브러리 내 함수들에 대한 정보

<br>

c/cpp 라이브러리
1. 정적 라이브러리 : 컴파일 시에 결합하는 방식
    - .lib : 현재는 기본적인 정보만 저장
2. 동적 라이브러리 : 실행 시에 결합하는 방식
    - .so (shared object, Unix/Linux)또는 ."dll"(dynamic-link library, Window)
    
<br>

OpenGL 라이브러리
- OpenGL core library
    - libGL.so : 대부분의 유닉스/리눅스 시스템, 맥 시스템
    - opengl32.dll : 마이크로 소프트 윈도우
        - 실제 내용은 64비트 cpu에 최적화된 업데이터 된 내용이 들어있음
        - 윈도우 그래픽스 드라이버를 업데이트 하면 opengl32.dll도 최신 버전으로 업데이트 됨
        
<br>

윈도우 시스템의 역할
- UI가 그래픽스 인터페이스로 제공되고 있음
- 윈도우 시스템의 역할
    - 2D 윈도우를 만들고 이동/크기조정/축소/삭제
    - 윈도우 내에 2D/3D 그래픽 출력은 OpenGL같은 컴퓨터 그래픽스 라이브러리에서 해야할 일

윈도우 시스템과 OpenGL과의 연결을 해주는 라이브러리는 각 윈도우 시스템마다 별도로 제공 됨
- WGL for MS-Windows
- 단일-플랫폼에서 OpenGL 프로그램은 해당 윈도우 연결 라이브러리만 사용하면 되니 문제 없다 OpenGL + WGL + MS-Win
- 멀티-플랫폼에서 OpenGL 프로그램을 지원하려면?


GLUT(OpenGL Utility Toolkit)
- 거의 모든 윈도우 시스템의 공통 기능만 제공
    - 운영체제들이 공통적으로 사용하는 기능을 모아서 하나의 라이브러리로 통합해서 제공해주자
- 장점 : 높은 이식성 portability
- 단점 : 제한된 기능
    - 특별한 윈도우 시스템에 있는 기능들은 제공하지 않는다.
- 관리 부재
    - 저작권 문제로, 업데이트 불가 
    - 새로운 버전과 작동 불능
    
freeglut
- GLUT의 독립적 구현

<br>

### GLFW(OpenGL Frame Work)
- open source, multi-platfrom 라이브러리
    - for OpenGL, OpenGL ES, Vulkan
    - 윈도우, 맥, x윈도우 시스템 모두 지원
    
- 모든 윈도우 시스템의 공통 기능 제공
    - 윈도우 생성/제거
    - 마우스/키보드 입력
    - 이벤트 처리

    

## OpenGL Extensions
어떤 버전에서든지 반드시 제공해야 하는 내용은 코어 라이브러리에서 제공  
핵심 기능은 아니지만 많은 사람들이 사용하고 있기에 표준화 시켜놓은 것을 OpenGL Extensions라는 이름으로 제공하고 있다.

- Why extensions?
    - 하드웨어 제공자가 좋은 기능을 바로 추가
    - 일단 extension 으로 등록 >> 나중에 core liblary 기능으로 포함 가능

<br>

OpenGL extension의 강점  
- core library만 사용한다면 높은 이식성(portability) 확보 가능
- extension을 사용하면 >> 최신 기능을 더 효과적으로 구현 가능
> e.g., AMD와 NVIDIA는 모두 그래픽 카드에 코어 라이브러리는 반드시 구현해서 제공하고 있다. 이식성을 강조하는 프로그램은 코어 라이브러리만 이용해서 프로그램을 짠다, 어떤 경우에든지 AMD 카드에도 NVIDIA 카드에도 돌아간다. 최신형 그래픽 카드에는 어떤 기능이 익스텐션으로 들어가있는 것을 발견 그러면 이 경우에는 해당 그래픽 카드를 사서 프로그램을 짜면 더 빨리 돌아감 하지만 다른 그래픽 카드에서는 돌아가지 않음

<br>

### GLEW(Open Extension Wrangler Library)_글루
- 현재 그래픽 카드에서 어떤 익스텐션을 사용할 수 있는지를 알려준다.

```c
# include <grew.h>
... // 필요한 자료구조, 함수 정의
glewInit(); // 반드시 실행!
if (glewIsSupported("GL_VERSION_4_0")){ // 문자열을 잡아둔다
    /*OpenGL 4,0버전 features 사용*/
}else{
    /*오픈지엘 구 버전 사용*/
}

if(glewIsSupported("GL_ARB_point_sprite")){
    /* 해당 extension features 사용 */
}else{
    /* 해당 extension 없이 구현 */
}

```

## OpenGL 라이브러리 특성

3D 그래픽스에서 필요한 모든 기능을 OpenGL 라이브러리가 모두 제공한다.
- e.g., 
    - OpenGL 담당: primitives 출력, 속성 설정, query, 좌표 변환
    - GLFW : 윈도우 제어, 장치 입력(키보드, 마우스)
    
OpenGL state 관리
- OpenGL은 state machine 개념으로 구현
    - 상태를 바꾸면 같은 삼각형도 다른식으로 그려질 수 있다.
- OpenGL draw 과정
    - 내부 상태 설정 >> 프리미티브 출력
    
객체 지향(Object orientation) 개념의 결여
- 객체 지향 개념이 생기기 전에 OpenGL 개념은 정립됨(1980s)
- function overloading이 지원되지 않아 사용하려는 자료형에 따라, 함수 이름이 바뀜


OpenGL 자료형, constants
```cpp
#include<GL/glew.h>
#include<GLFW/glfw3.h>
// 실제로는 <GL/gl.h>, <GLES/gles2.h> 등에 들어있지만, 
// 위의 헤더 파일들이 자동으로 가져옴
```
